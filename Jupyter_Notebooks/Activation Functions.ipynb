# Activation Functions

```python
import numpy as np
from math import exp
import matplotlib.pyplot as plt
```

First we create a function for each of the activation functions.

```python
def sigmoid(x):
    y = np.exp(x)/(np.exp(x)+1)
    return y

def relu(x):
    return np.maximum(0,x)

def tanh(x):
    y = (np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x))
    return y
```

# Sigmoid

```python
X = np.arange(-6, 7, 0.1)

Y1 = sigmoid(X)

plt.plot(X, Y1)

plt.show()
```

# ReLU

```python
Y2 = relu(X)
plt.plot(X, Y2)
plt.show
```

# Tanh

```python
Y3 = tanh(X)
plt.plot(X, Y3)
plt.show
```
